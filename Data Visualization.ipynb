# Plotting Bivariate Distributions

## Hexplot

```python
sns.jointplot(x="engine_size", y="bore", data=df);
```

```python
##Hexplot is a bivariate analog of histogram as it shows the number of observations that falls within hexagonal bins. This is a plot which works with large dataset very easily. To draw a hexplot, we’ll set kind attribute to hex.
sns.jointplot(x=df.bore, y=df.engine_size, kind="hex", data = df)
```

## KDE Plot using Seaborn

```python
##achieved by using jointplot() function and setting the attribute kind to KDE.
sns.jointplot(x="bore", y="engine_size", data=df, kind="kde");
```

## Heatmaps using Seaborn

```python
##Heatmaps are graphical representations in which each variable is represented as a color.
corrmat = df.corr()
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(corrmat, vmax=.8, square=True)
```

## Boxen Plot using Seaborn

```python
## Boxen plots were originally named letter value plot as it shows large number of values of a variable, also known as quantiles. These quantiles are also defined as letter values. By plotting a large number of quantiles, provides more insights about the shape of the distribution. These are similar to box plots
sns.catplot(x="bore", y="engine_size", data=df, kind="boxen",height=4, aspect=2.7, hue = "aspiration")
```

## Ridge Plot using seaborn

```python
##It is also called joyplot. Ridge plot helps in visualizing the distribution of a numeric value for several groups. These distributions could be represented by using KDE plots or histograms.
# Initialize the FacetGrid object
g = sns.FacetGrid(df, row="aspiration", hue="aspiration", aspect=5, height=3)

# # Draw the densities in a few steps
g.map(sns.kdeplot, "bore", shade=True, alpha=1, lw=3.5, bw=.2)
g.map(sns.kdeplot, "bore", color="w", lw=2, bw=.2)
g.map(plt.axhline, y=0, lw=2)

# # Define and use a simple function to label the plot in axes coordinates
def label(x, color, label):
    ax = plt.gca()
    ax.text(0, .2, label, color=color, ha="left", va="center", transform=ax.transAxes)
g.map(label, "bore")

# Set the subplots to overlap
g.fig.subplots_adjust(hspace=-.25)

# # Remove axes details that don't play well with overlap
g.set_titles("")
g.set(yticks=[])
g.despine(bottom=True, left=True)
```
